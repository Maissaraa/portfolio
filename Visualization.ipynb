## Visualization 5: Count Plot - Sales by Month:


## Visualization 7: Line Plot - Monthly Sales Trend: 
Displaying the monthly sales trend over time using a line plot.

```python
monthly_sales = E_comm_USA.resample('M', on='Ship Date')['Sales'].sum() 
monthly_sales
```

```python
# Visualization 7: Line Plot - Monthly Sales Trend 
plt.plot(monthly_sales,marker='o') 
plt.xlabel('Date') 
plt.ylabel('Monthly Sales') 
plt.xticks(rotation=45)
plt.title('Monthly Sales Trend') 
plt.show()
```

```python
monthly_dicounts = E_comm_USA.resample('Y', on='Ship Date')['Discount'].count() 
# monthly_dicounts
# Visualization 7: Line Plot - Monthly Sales Trend 
plt.plot(monthly_dicounts,marker='o') 
plt.xlabel('Date') 
plt.ylabel('monthly_dicounts') 
plt.xticks(rotation=45)
plt.title('monthly_dicounts Trend') 
plt.show()
```
